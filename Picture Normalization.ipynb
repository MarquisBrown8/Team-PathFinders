```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg


image_files = ['Leaf_Image.jpeg', 'Leaf_Image2.jpeg', 'Leaf_Image3.jpeg']
resized_images = []


for file in image_files:
    img = mpimg.imread(file)  # Read image
    plt.imshow(img)  # Display original image
    plt.axis('off')
    plt.show()

    resized = cv2.resize(img, (224, 224))  # Resize
    resized = cv2.cvtColor(resized, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB
    resized = resized / 255.0  # Normalize pixel values

    resized_images.append(resized)
    print(f"{file} resized and normalized:", resized.shape)


resized_images = np.array(resized_images)
```
